Open files for cleaning

```python
df = pd.read_excel('CustomerCallList.xlsx')

newdf = df.copy()




# copy()



# isna()
# dropna()
# is_duplicated()
# drop_duplicates()
# strip()
# replace()
# to_string()
```

cleaning Last_name using the strip method

```python
newdf['Last_Name'] = newdf['Last_Name'].str.strip('/..*_')

newdf
```

Droping Not_useful_Column using drop method

```python
newdf = newdf.drop(columns =['Not_Useful_Column'])
newdf
```

Rearranging Phone Number Column using 

```python
newdf['Phone_Number'] = newdf['Phone_Number'].apply(lambda z: str(z))


newdf['Phone_Number'] = newdf['Phone_Number'].str.replace('[^ZA-Z0-9]', '', regex = True)



newdf['Phone_Number'] = newdf['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:9])


newdf['Phone_Number'] = newdf['Phone_Number'].str.replace('--', '')


newdf['Phone_Number'] = newdf['Phone_Number'].str.replace('N', '')



newdf
```

Splitting Address column into Three Columns

```python
newdf[['Street Address', 'State', 'Zip Code']] = newdf['Address'].str.split(',', n=2, expand = True)


newdf = newdf.drop(columns = 'Address')

newdf
```

```python
newdf['Paying Customer'] = newdf['Paying Customer'].str.replace('No', 'N')

newdf['Paying Customer'] = newdf['Paying Customer'].str.replace('N', 'No')

newdf['Paying Customer'] = newdf['Paying Customer'].str.replace('Yes', 'Y')

newdf['Paying Customer'] = newdf['Paying Customer'].str.replace('Y', 'Yes')

newdf['Paying Customer'] = newdf['Paying Customer'].str.replace('No/a', 'No')

newdf


```

```python
newdf['Do_Not_Contact'] = newdf['Do_Not_Contact'].str.replace('No','N')

newdf['Do_Not_Contact'] = newdf['Do_Not_Contact'].str.replace('N','No')

newdf['Do_Not_Contact'] = newdf['Do_Not_Contact'].str.replace('Yes','Y')

newdf['Do_Not_Contact'] = newdf['Do_Not_Contact'].str.replace('Y','Yes')

# newdf['Do_Not_Contact'] = newdf['Do_Not_Contact'].str.fillna('')

newdf




```

Droping all the NaN Value

```python
for x in newdf.index:
    if newdf.loc[x, 'Do_Not_Contact'] == 'Yes':
        newdf.drop(x, inplace = True)
        
newdf
```

```python

newdf = newdf.fillna('')
newdf
```
